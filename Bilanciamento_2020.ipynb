```python
############# BILANCIAMENTO 5 CLASSI PE IMMAGINI ################

import numpy as np
import pickle
import os

# Parametri

# Percorsi
directory = r"C:\Users\user\Documents\TESI\datigrezzi\DATI_grezzi_2020\Dati_ML_plus_5classi"
output_dir = os.path.join(directory, "bilanciamento_5classi")
os.makedirs(output_dir, exist_ok=True)

with open(os.path.join(directory, 'df_features.pickle'), 'rb') as f:
    df_features = pickle.load(f)
with open(os.path.join(directory, 'DPR_ml.pickle'), 'rb') as f:
    DPR = pickle.load(f)
with open(os.path.join(directory, 'maschera_giorno_ml.pickle'), 'rb') as f:
    GIORNO_maschera = pickle.load(f)

data_dict = {
    'df_features': df_features,
    'DPR_ml': DPR,
    'maschera_giorno_ml': GIORNO_maschera
}

max_value = np.max(DPR)
rain_intervals = [
    (0, 0.1),
    (0.1, 1),
    (1, 5),
    (5, 15),
    (15, max_value),
]

# Identificazione indici per ciascuna classe
balanced_indices = []
min_samples = float('inf')
class_indices = []

for (low, high) in rain_intervals:
    idx = np.where((DPR >= low) & (DPR < high))[0]
    class_indices.append(idx)
    min_samples = min(min_samples, len(idx))

# Sottocampionamento uniforme
for idx in class_indices:
    selected = np.random.choice(idx, size=min_samples, replace=False)
    balanced_indices.append(selected)

balanced_indices = np.sort(np.concatenate(balanced_indices))

# Selezione dei dati bilanciati
balanced_data = {}
for key, array in data_dict.items():
    if hasattr(array, 'iloc'):
        balanced_data[key] = array.iloc[balanced_indices].reset_index(drop=True)
    else:
        balanced_data[key] = array[balanced_indices]


# Salvataggio
for key, array in balanced_data.items():
    with open(os.path.join(output_dir, f"{key}.pickle"), 'wb') as f:
        print(f"Chiave salvata: {key}.pickle")
        pickle.dump(array, f)

print("✔️ Dati bilanciati per pixel salvati.")
```
