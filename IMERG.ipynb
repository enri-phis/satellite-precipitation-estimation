```python
################# MAPPE ISTANTANEE #####################

import pickle
import numpy as np
import matplotlib.pyplot as plt
import os
import cartopy.crs as ccrs
import cartopy.feature as cfeature
import matplotlib
import pandas as pd

# === Percorsi ===
output_file = r"D:\Enrico\definitivo_actual\y_pred.npy"
dir_lat = r"D:\Enrico\definitivo_actual\Dati_processati\lat.pickle"
dir_lon = r"D:\Enrico\definitivo_actual\Dati_processati\lon.pickle"
dir_time = r"D:\Enrico\definitivo_actual\Dati_processati\TIME.pickle"
dir_giorno = r"D:\Enrico\definitivo_actual\Dati_processati\giorno_maschera_70.pickle"
output_dir = r"D:\Enrico\definitivo_actual\Mappe_SEVIRI_istantanee"
os.makedirs(output_dir, exist_ok=True)

# === Carico lat, lon e time ===
with open(dir_lat, "rb") as f:
    lat_all = pickle.load(f)
with open(dir_lon, "rb") as f:
    lon_all = pickle.load(f)
with open(dir_time, "rb") as f:
    time_all = pickle.load(f)
with open(dir_giorno, "rb") as f:
    GIORNO_maschera = pickle.load(f)

# === Applico maschera giorno ===
mask_indices_bal = np.where(GIORNO_maschera == 1)[0]
time_all = np.array(time_all)[mask_indices_bal]

# === Carico predizioni ===
y_pred = np.load(output_file)

# === Parametri immagine ===
img_h, img_w = 55, 112
pixels_per_image = img_h * img_w
num_images = len(y_pred) // pixels_per_image
print(f"Totale immagini: {num_images}")

# === Definizione colori ===
colors = [
    '#FFFFFF',  # Dry
    '#A4D3EE',  # Light blue
    '#32CD32',  # Green
    '#FFD700',  # Gold
    '#8A2BE2'   # Violet
]
cmap = matplotlib.colors.ListedColormap(colors, N=5)

# === Limiti geografici ===
lat_min, lat_max = 43.4, 45.8
lon_min, lon_max = 9.15, 13.2

# === Loop su ogni immagine (istantanea) ===
for i in range(num_images):
    start = i * pixels_per_image
    end = (i + 1) * pixels_per_image

    cls_map = y_pred[start:end]
    img_data = cls_map.reshape(img_h, img_w)
    lat_2d = lat_all[:pixels_per_image].reshape(img_h, img_w)
    lon_2d = lon_all[:pixels_per_image].reshape(img_h, img_w)

    # Correggi orientamento
    img_data = np.fliplr(img_data)

    # Timestamp
    t_pixel = pd.to_datetime(time_all[start])
    t_str = t_pixel.strftime("%Y-%m-%d %H:%M")
    t_save = t_pixel.strftime("%Y%m%d_%H%M")

    # === Plot ===
    fig = plt.figure(figsize=(8,6))
    ax = plt.axes(projection=ccrs.PlateCarree())

    ax.set_extent([lon_2d.min(), lon_2d.max(), lat_2d.min(), lat_2d.max()], crs=ccrs.PlateCarree())
    im = ax.imshow(img_data, origin='lower', cmap=cmap,
                   extent=[lon_2d.min(), lon_2d.max(), lat_2d.min(), lat_2d.max()],
                   transform=ccrs.PlateCarree(), vmin=0, vmax=4)

    # Griglia e confini
    xticks = np.linspace(lon_min, lon_max, 5)[1:-1]
    yticks = np.linspace(lat_min, lat_max, 5)[1:-1]
    ax.set_xticks(xticks)
    ax.set_yticks(yticks)

    ax.add_feature(cfeature.COASTLINE)
    ax.add_feature(cfeature.BORDERS, linestyle=':')
    ax.add_feature(cfeature.LAND, facecolor='lightgray', alpha=0.3)

    # Colorbar
    cbar = plt.colorbar(im, ax=ax, orientation='vertical', pad=0.04, shrink=0.4)
    cbar.set_label('Precipitation class (instantaneous)', fontsize=13)
    cbar.set_ticks([0.5, 1.5, 2.5, 3.5, 4.5])
    cbar.set_ticklabels(['Dry', 'Light', 'Moderate', 'Heavy', 'Intense'])
    cbar.ax.tick_params(labelsize=10)

    ax.set_xlabel('Longitude [°]', fontsize=12)
    ax.set_ylabel('Latitude [°]', fontsize=12)
    #ax.set_title(f"Instantaneous precipitation - {t_str}", fontsize=12)

    ax.tick_params(labelsize=10, direction='out', length=3, width=0.8)
    ax.xaxis.set_tick_params(pad=4)
    ax.yaxis.set_tick_params(pad=4)

    save_path = os.path.join(output_dir, f"precip_{t_save}.png")
    plt.tight_layout()
    plt.savefig(save_path, dpi=150)
    plt.close()
    print(f"✅ Mappa salvata: {save_path}")

print("🎉 Tutte le mappe istantanee generate correttamente!")
```
