```python
########### conversione #############

import os, zipfile
import xarray as xr
from satpy import Scene
import warnings
warnings.filterwarnings("ignore")

chan = [
    'VIS006', 'VIS008','IR_016', 'IR_039', 'WV_062', 'WV_073',
    'IR_087', 'IR_097', 'IR_108', 'IR_120', 'IR_134'
]

def nat2net(sdir, f):
    zip_path = os.path.join(sdir, f)
    with zipfile.ZipFile(zip_path) as zf:
        fnat = next((name for name in zf.namelist() if name.endswith('.nat')), None)
        if fnat is None:
            print(f"[SKIP] Nessun .nat trovato in {f}")
            return None
        zf.extract(fnat, sdir)  # estrai direttamente nella cartella dei dati

    nat_path = os.path.join(sdir, fnat)
    print(f"[INFO] Elaboro {nat_path}")

    scn = Scene([nat_path], reader='seviri_l1b_native', reader_kwargs={'fill_disk': False})
    scn.load(chan)

    ds = {}
    for k, c in enumerate(chan):
        ds[f'channel_{k+1}'] = (('x', 'y'), scn[c].values)

    ds = xr.Dataset(ds, coords={
        'x': scn['VIS006'].coords['x'].values,
        'y': scn['VIS006'].coords['y'].values
    })
    ds.attrs = scn[chan[0]].attrs['orbital_parameters']

    out_file = os.path.join(sdir, f.replace(".zip", ".nc"))
    ds.to_netcdf(out_file)
    print(f"[OK] Salvato {out_file}")

    # pulizia
    os.remove(nat_path)
    return out_file


def batch_convert(source):
    for f in os.listdir(source):
        if f.endswith(".zip"):
            out_file = os.path.join(source, f.replace(".zip", ".nc"))
            # 👇 qui c’è il controllo
            if os.path.exists(out_file):
                print(f"[SKIP] {out_file} esiste già, salto.")
                continue
            try:
                nat2net(source, f)
            except Exception as e:
                print(f"[ERROR] Errore durante l'elaborazione di {f}: {e}")
        else:
            print(f"[DEBUG] Saltato: {f} (non è un file .zip)")



if __name__ == "__main__":
    data_directory = r"D:\\Enrico\\Dati_eumetsat"
    print(f"[DEBUG] Inizio elaborazione in {data_directory}")
    batch_convert(data_directory)
    print("[DEBUG] Elaborazione completata")
```

```python
import xarray as xr
import numpy as np
from pyproj import Proj
import os
import re
from datetime import datetime

def crop_to_region(nc_input_path, nc_output_path,
                   lat_min, lat_max, lon_min, lon_max):
    """
    Ritaglia un file NetCDF alla regione definita da lat_min, lat_max, lon_min, lon_max.
    Controlli di debug inclusi per garantire la coerenza tra dati e coordinate.
    """
    try:
        # --- Apro dataset ---
        ds = xr.open_dataset(nc_input_path)
        print(f"\n📂 Apertura file: {nc_input_path}")
        print(f"Variabili nel dataset: {list(ds.data_vars)}")

        proj_geo = Proj(proj='geos',
                 h=ds.attrs.get('satellite_nominal_altitude', 35784322.93211132),
                 lon_0=ds.attrs.get('satellite_nominal_longitude', 0.0),
                 lat_0=ds.attrs.get('satellite_nominal_latitude', 0.0),
                 sweep='y',
                 a=6378169.0, b=6356583.8)
        # --- Coordinate originali ---
        x = ds['x'].values
        
        y = ds['y'].values
        X, Y = np.meshgrid(x, y)
        lon, lat = proj_geo(X, Y, inverse=True)
        
        print(f"Shape originale x: {x.shape}, {y.shape}")
        print(f"Shape originale X: {X.shape}, {Y.shape}")
        print(f"Shape originale lat/lon: {lat.shape}, {lon.shape}")
        print(f"Lat min/max: {lat.min()}/{lat.max()}, Lon min/max: {lon.min()}/{lon.max()}")

        # --- Maschera regione ---
        mask = (lat >= lat_min) & (lat <= lat_max) & (lon >= lon_min) & (lon <= lon_max)
        ys, xs = np.where(mask)
        if len(xs) == 0 or len(ys) == 0:
            print(f"⚠️ Nessun pixel trovato per {os.path.basename(nc_input_path)}")
            ds.close()
            return

        x0, x1 = xs.min(), xs.max() + 1
        y0, y1 = ys.min(), ys.max() + 1
        cropped = ds.isel(x=slice(x0, x1), y=slice(y0, y1))

        print(f"Pixel ritagliati: x={x0}:{x1}, y={y0}:{y1}")
        print(f"Shape dati ritagliati: {[v.shape for v in cropped.data_vars.values()]}")

    
        lat_cropped = lat[y0:y1, x0:x1] # senza trasposizione
        lon_cropped = lon[y0:y1, x0:x1]
        print(f"Shape lat/lon cropped: {lat_cropped.shape}, {lon_cropped.shape}")

        # Debug coerenza valori
        print(f"Lat sample: {lat_cropped.flatten()[:5]}, Lon sample: {lon_cropped.flatten()[:5]}")

        cropped = cropped.assign_coords({
            'lat': (('y','x'), lat_cropped),
            'lon': (('y','x'), lon_cropped)
        })

        # --- Estrazione timestamp dal nome file ---
        filename = os.path.basename(nc_input_path)
        match = re.search(r"(\d{14})", filename)  # prende YYYYMMDDHHMMSS
        timestamp = datetime.strptime(match.group(1), "%Y%m%d%H%M%S") if match else None
        if timestamp:
            cropped = cropped.assign_coords(time=("time", [timestamp]))
        else:
            print(f"⚠️ Impossibile estrarre timestamp da {filename}")

        # --- Salvataggio con debug ---
        try:
            cropped.to_netcdf(nc_output_path)
            print(f"✅ Salvato file ritagliato: {nc_output_path} con timestamp {timestamp}")
        except Exception as e:
            print(f"⚠️ Errore nel salvare {nc_output_path}: {e}")

    finally:
        ds.close()
        if 'cropped' in locals():
            cropped.close()


if __name__ == "__main__":
    # --- Percorsi ---
    input_folder = r"D:\\Enrico\\Dati_convertiti"
    output_folder = r"D:\\Enrico\\NUOVO"
    os.makedirs(output_folder, exist_ok=True)

    # --- Limiti regione Emilia-Romagna ---
    lat_min, lat_max = 43.4, 45.8  ##### per imrg meto 46
    lon_min, lon_max = 9.15, 13.2

    for file_name in os.listdir(input_folder):
        if not file_name.endswith(".nc"):
            continue

        input_path = os.path.join(input_folder, file_name)
        output_path = os.path.join(output_folder, f"ER_{file_name}")

        # --- SALTO FILE GIA' PROCESSATI ---
        if os.path.exists(output_path):
            print(f"⏭️ File già elaborato, salto: {file_name}")
            continue

        try:
            crop_to_region(input_path, output_path, lat_min, lat_max, lon_min, lon_max)
        except Exception as e:
            print(f"⚠️ Errore nel processare {file_name}, salto: {e}")
            continue
        
        print(f"✅ Tutti i file NetCDF processati e salvati (o saltati se errore) in {output_folder}")

    
```

```python
############# maschera stagioni ##########


import os
import numpy as np
import xarray as xr
import random
import pandas as pd

# ---------------------- CONFIGURAZIONE CARTELLE ----------------------
data_directory = r"D:\\Enrico\\emiliaromagna"
mask_directory = r"D:\\Enrico\\emiliaromagna\\MASCHERE"
os.makedirs(mask_directory, exist_ok=True)

# ---------------------- FUNZIONE STAGIONALE ----------------------
def get_season_by_day_of_year(day_of_year):
    """
    Restituisce:
    0 -> Estate
    1 -> Inverno
    2 -> Primavera/Autunno (da escludere)
    """
    if 335 <= day_of_year <= 365 or 1 <= day_of_year <= 59:
        return 1  # Inverno
    elif 152 <= day_of_year <= 243:
        return 0  # Estate
    else:
        return 2  # Primavera/Autunno

# ---------------------- ELABORAZIONE ----------------------
nc_files = sorted([f for f in os.listdir(data_directory) if f.endswith(".nc")])
if not nc_files:
    print("⚠️ Nessun file .nc trovato!")
else:
    print(f"📂 Trovati {len(nc_files)} file .nc da elaborare.")

# Lista finale per la maschera stagionale
stagionalita_maschera = []

for file_name in nc_files:
    file_path = os.path.join(data_directory, file_name)
    print(f"\n📄 Elaboro file: {file_name}")

    try:
        ds = xr.open_dataset(file_path)

        # Estrazione delle variabili lat, lon e time
        lat = ds['lat'].values.flatten()
        lon = ds['lon'].values.flatten()

        # Estrazione del timestamp
        timestamp = ds['time'].values[0]  # datetime64[ns] per il primo (o unico) timestep
        day_of_year = pd.to_datetime(str(timestamp)).dayofyear  # converte in giorno dell'anno

        print(f"🔍 Prima latitudine: {lat[0]}, Prima longitudine: {lon[0]}")
        sample_indices = random.sample(range(len(lat)), min(5, len(lat)))
        print(f"🗺️ Elaboro {len(lat)} punti geografici...")

        for i, (la, lo) in enumerate(zip(lat, lon)):
            try:
                stagione = get_season_by_day_of_year(day_of_year)
                stagionalita_maschera.append(stagione)
                if i in sample_indices:
                    print(f"🌍 Lat: {la}, Lon: {lo}, Stagione: {stagione}")
            except Exception as e:
                print(f"⚠️ Errore con lat={la}, lon={lo}: {e}")
                stagionalita_maschera.append(2)

        ds.close()

    except Exception as e:
        print(f"❌ Errore nel caricamento di {file_name}: {e}")

# ---------------------- SALVATAGGIO ----------------------
print("\n💾 Salvataggio della maschera stagionale in formato .mat...")
from scipy.io import savemat
stagionalita_dict = {"stagionalita_maschera": np.array(stagionalita_maschera)}
output_path = os.path.join(mask_directory, 'stagionalita_maschera_estate_inverno.mat')
savemat(output_path, stagionalita_dict)
print(f"✅ Maschera stagionale salvata con successo in '{output_path}'!")
```

```python
########## maschera mare-terra ############

import os
import numpy as np
import xarray as xr
from scipy.interpolate import griddata
import scipy.io as sio

# ---------------------- CONFIGURAZIONE CARTELLE ----------------------
data_directory =  r"D:\\Enrico\\emiliaromagna"
mask_directory =  r"D:\\Enrico\\emiliaromagna\\MASCHERE"
os.makedirs(mask_directory, exist_ok=True)

def carica_sealand_mask(netcdf_file):
    """Carica la maschera Mare/Terra dal file NetCDF."""
    ds = xr.open_dataset(netcdf_file)
    
    # Usa i nomi effettivi nel file NetCDF della maschera
    latitudes = ds['latitude'].values
    longitudes = ds['longitude'].values
    sealand_mask = ds['lsm'].values[0,:,:]  # maschera LSM

    ds.close()
    return latitudes, longitudes, sealand_mask

def regridding(latitudes_dati, longitudes_dati, mask, latitudes_mask, longitudes_mask):
    """Interpolazione della maschera sui dati."""
    grid_x, grid_y = np.meshgrid(longitudes_mask, latitudes_mask)
    points = np.array([grid_x.flatten(), grid_y.flatten()]).T
    mask_values = mask.flatten()
    regridded = griddata(points, mask_values, (longitudes_dati, latitudes_dati), method='nearest')
    return regridded

def salva_mascherati(latitudes, longitudes, mare_mask, terra_mask, output_dir):
    """Salva le maschere Mare e Terra come file .mat."""
    sio.savemat(f'{output_dir}/Mare_mask_regridded.mat', {'mask': mare_mask, 'latitudes': latitudes, 'longitudes': longitudes})
    sio.savemat(f'{output_dir}/Terra_mask_regridded.mat', {'mask': terra_mask, 'latitudes': latitudes, 'longitudes': longitudes})

# ---------------------- RACCOLTA LAT/LON DA TUTTI I FILE ----------------------
latitudes_list, longitudes_list = [], []

nc_files = sorted([f for f in os.listdir(data_directory) if f.endswith(".nc")])
for file_name in nc_files:
    file_path = os.path.join(data_directory, file_name)
    ds = xr.open_dataset(file_path)
    latitudes_list.append(ds['lat'].values.flatten())
    longitudes_list.append(ds['lon'].values.flatten())
    ds.close()

latitudes_dati = np.concatenate(latitudes_list)
longitudes_dati = np.concatenate(longitudes_list)

# ---------------------- CARICAMENTO DELLA MASCHERA ----------------------
netcdf_mask_file = 'C:/Users/user/Documents/TESI/datigrezzi/sea_land.nc'
latitudes_mask, longitudes_mask, sealand_mask = carica_sealand_mask(netcdf_mask_file)

# ---------------------- REGRIDDING ----------------------
regridded_mask = regridding(latitudes_dati, longitudes_dati, sealand_mask, latitudes_mask, longitudes_mask)

# ---------------------- CREAZIONE MASCHERE MARE/TERRA ----------------------
mare_mask = (regridded_mask == 0).astype(int)
terra_mask = (regridded_mask == 1).astype(int)

# ---------------------- SALVATAGGIO ----------------------
salva_mascherati(latitudes_dati, longitudes_dati, mare_mask, terra_mask, mask_directory)
print("Maschere Mare e Terra salvate con successo!")
```

```python
import os
import numpy as np
import pickle
import re
import xarray as xr
import scipy.io as sio
from datetime import datetime, timedelta

# === Percorsi cartelle ===
cartella_base = r"D:\\Enrico\\emiliaromagna"
cartella_dati = os.path.join(cartella_base)
cartella_output = os.path.join(cartella_base, "Dati_processati")
cartella_maschere = os.path.join(cartella_base, "MASCHERE")

# Crea cartella di output se non esiste
os.makedirs(cartella_output, exist_ok=True)

# Dizionari per concatenare i dati
dati_concatenati = {}
time_concatenato = []
lat_concatenato = []
lon_concatenato = []

# Funzione per ordinare i file in base a DOY e TIME se presenti nel nome
def estrai_doy_time(nome_file):
    match = re.search(r"DOY(\d+)_TIME(\d+)", nome_file)
    if match:
        return int(match.group(1)), int(match.group(2))
    return float('inf'), float('inf')

# Lista file NetCDF ordinata
file_ordinati = sorted(
    [f for f in os.listdir(cartella_dati) if f.endswith('.nc')],
    key=lambda x: estrai_doy_time(x)
)

print("Caricamento e concatenazione dei dati NetCDF...")

for filename in file_ordinati:
    filepath = os.path.join(cartella_dati, filename)
    ds = xr.open_dataset(filepath)

    # --- Variabili data_vars (canali) ---
    for chiave in ds.data_vars:
        valore = ds[chiave].values.flatten()
        if chiave not in dati_concatenati:
            dati_concatenati[chiave] = [valore]
        else:
            dati_concatenati[chiave].append(valore)

    # --- Tempo per ogni pixel ---
    if "time" in ds.coords:
        t_values = np.broadcast_to(ds["time"].values, (np.prod(list(ds.sizes.values())),))
        time_concatenato.append(t_values)

    # --- Lat/Lon per ogni pixel ---
    if "lat" in ds.coords:
        lat_concatenato.append(ds["lat"].values.T.flatten())  # Trasposta per allineare correttamente
    if "lon" in ds.coords:
        lon_concatenato.append(ds["lon"].values.T.flatten())

    ds.close()

# --- Concatenazione finale dei canali con debug ---
for chiave in dati_concatenati:
    dati_array = np.concatenate(dati_concatenati[chiave]).ravel()
    match = re.match(r'(?:CH|channel)[ _]?(\d+)', chiave, re.IGNORECASE)
    chiave_pickle = f"CH_{match.group(1)}" if match else chiave

    # --- DEBUG: primi valori canale ---
    print(f"\n📌 Debug dati canale {chiave_pickle}:")
    print("Primi 10 valori del canale:", dati_array[:10])
    print("Ultimi 10 valori del canale:", dati_array[-10:])

    with open(os.path.join(cartella_output, f"{chiave_pickle}.pickle"), 'wb') as f:
        pickle.dump(dati_array, f)
    print(f"✅ {chiave_pickle} salvato, shape: {dati_array.shape}")

# --- Concatenazione TIME ---
time_array = np.concatenate(time_concatenato).ravel()
TIME_dt = [t.astype('O') if isinstance(t, np.datetime64) else t for t in time_array]

print("\n📌 Debug TIME:")
print("Primi 10 valori TIME:", TIME_dt[:10])
print("Ultimi 10 valori TIME:", TIME_dt[-10:])

with open(os.path.join(cartella_output, "TIME.pickle"), "wb") as f:
    pickle.dump(TIME_dt, f)
print(f"✅ TIME salvato, elementi: {len(TIME_dt)}")

# --- Concatenazione lat/lon con debug ---
lat_all = np.concatenate(lat_concatenato).ravel()
lon_all = np.concatenate(lon_concatenato).ravel()

print("\n📌 Debug lat/lon:")
print("Primi 10 valori lat:", lat_all[:10])
print("Primi 10 valori lon:", lon_all[:10])
print("Ultimi 10 valori lat:", lat_all[-10:])
print("Ultimi 10 valori lon:", lon_all[-10:])

# --- Salvataggio lat/lon ---
with open(os.path.join(cartella_output, "lat.pickle"), "wb") as f:
    pickle.dump(lat_all, f)
print(f"✅ lat salvato, shape: {lat_all.shape}")

with open(os.path.join(cartella_output, "lon.pickle"), "wb") as f:
    pickle.dump(lon_all, f)
print(f"✅ lon salvato, shape: {lon_all.shape}")

# --- Salvataggio maschere .mat ---
file_maschere = [f for f in os.listdir(cartella_maschere) if f.endswith('.mat')]
for file_maschera in file_maschere:
    path_maschera = os.path.join(cartella_maschere, file_maschera)
    mat_data = sio.loadmat(path_maschera)
    
    chiavi_utili = [k for k in mat_data.keys() if not k.startswith('__')]
    if chiavi_utili:
        maschera = mat_data[chiavi_utili[0]].ravel()
        nome_maschera = os.path.splitext(file_maschera)[0]
        with open(os.path.join(cartella_output, f"{nome_maschera}.pickle"), 'wb') as f:
            pickle.dump(maschera, f)
        print(f"✅ Maschera {nome_maschera} salvata, shape: {maschera.shape}")
    else:
        print(f"⚠️ Nessuna chiave utile trovata in {file_maschera}")

print("\n🎉 Tutti i file sono stati salvati correttamente!")
```

```python
############# creo le features (matrice di imput) ############

import os
import numpy as np
import pickle
import pandas as pd
from itertools import combinations
from scipy.ndimage import uniform_filter, laplace
from skimage.filters.rank import entropy
from skimage.morphology import disk
from skimage.util import img_as_ubyte

# === Percorsi ===
cartella_input = r"D:\Enrico\emiliaromagna\Dati_processati"
output_dir = r"D:\Enrico\emiliaromagna\Dati_features"
os.makedirs(output_dir, exist_ok=True)

# === File da caricare ===
file_names = [
    'CH_1.pickle', 'CH_2.pickle', 'CH_3.pickle', 'CH_4.pickle', 'CH_5.pickle',
    'CH_6.pickle', 'CH_7.pickle', 'CH_8.pickle', 'CH_9.pickle', 'CH_10.pickle', 'CH_11.pickle',
    'Mare_mask_regridded.pickle', 'Terra_mask_regridded.pickle', 'stagionalita_maschera_estate_inverno.pickle'
]

# === Caricamento dei pickle ===
for file_name in file_names:
    file_path = os.path.join(cartella_input, file_name)
    if os.path.exists(file_path):
        with open(file_path, 'rb') as f:
            nome_variabile = os.path.splitext(file_name)[0]
            globals()[nome_variabile] = pickle.load(f).ravel()  # appiattisce in 1D
            print(f"{nome_variabile} caricato, shape: {globals()[nome_variabile].shape}")
    else:
        print(f"{file_name} non trovato!")

# Rinominazione maschere per uniformità
Mare_maschera = globals()['Mare_mask_regridded']
TERRA_maschera = globals()['Terra_mask_regridded']
STAGIONI_maschera = globals()['stagionalita_maschera_estate_inverno']

# Funzioni per applicare maschere
def applica_maschera(canale, maschera, valore_maschera=1):
    return np.where(maschera == valore_maschera, canale, 0)

# Funzione per sostituire NaN con il valore valido più vicino
#def replace_nans_with_nearest(data):
#    valid_indices = np.flatnonzero(~np.isnan(data))
#    invalid_indices = np.flatnonzero(np.isnan(data))
#    for idx in invalid_indices:
#        closest_valid_index = valid_indices[np.abs(valid_indices - idx).argmin()]
#        data[idx] = data[closest_valid_index]
#    return data

def replace_nans_with_nearest_blockwise(arr, block_size=1_198_800):
    """
    Sostituisce NaN con il valore valido più vicino in array 1D, a blocchi
    per ridurre il consumo di memoria.
    """
    n = len(arr)
    result = np.empty_like(arr)
    for start in range(0, n, block_size):
        end = min(start + block_size, n)
        block = arr[start:end]
        result[start:end] = replace_nans_with_nearest(block)
        print(f"✅ Blocco {start}:{end} processato")
    return result

def replace_nans_with_nearest(arr):
    """
    Sostituisce i NaN con il valore valido più vicino lungo l'array 1D.
    Versione veloce e vectorizzata.
    """
    arr = arr.copy()
    n = len(arr)
    isnan = np.isnan(arr)
    notnan = ~isnan
    indices = np.arange(n)
    
    # Indici del valore valido più vicino a sinistra
    left = np.where(notnan, indices, 0)
    np.maximum.accumulate(left, out=left)
    
    # Indici del valore valido più vicino a destra
    right = np.where(notnan, indices, n-1)
    np.minimum.accumulate(right[::-1], out=right[::-1])
    
    # Sostituisci i NaN con il valore più vicino (sinistra o destra)
    nearest = np.where(isnan, np.where((indices - left) <= (right - indices), left, right), indices)
    arr[isnan] = arr[nearest[isnan]]
    
    return arr

# Flatten dei canali e DPR (già appiattiti in riga)
canali = [f'CH_{i}' for i in range(1, 12)]
for ch in canali:
    globals()[ch] = globals()[ch].ravel()

# === Media e std mobile ===
import numpy as np
from scipy.ndimage import uniform_filter

import numpy as np
from scipy.ndimage import uniform_filter

# === Funzione media/std mobile immagini irregolari ===
import numpy as np
from scipy.ndimage import uniform_filter

import numpy as np
from scipy.ndimage import uniform_filter

from scipy.ndimage import uniform_filter
import numpy as np

import numpy as np
from scipy.signal import convolve2d

def media_std_mobile_immagini(vettore_1D, width, height, kernel_size=3):
    """
    Calcola media e deviazione standard mobile 2D su immagini (h,w).
    Prima vanno rimossi i NaN dal vettore 1D.
    """
    num_pixel = width * height
    num_immagini = len(vettore_1D) // num_pixel
    vettore_media, vettore_std = [], []

    # Kernel uniforme
    kernel = np.ones((kernel_size, kernel_size), dtype=float)
    kernel /= kernel.size

    for i in range(num_immagini):
        # Ricostruisci immagine
        img = vettore_1D[i*num_pixel:(i+1)*num_pixel].reshape(width, height)

        # Media mobile con convoluzione
        media = convolve2d(img, kernel, mode='same', boundary='symm')

        # Deviazione standard mobile
        mean_sq = convolve2d(img**2, kernel, mode='same', boundary='symm')
        std = np.sqrt(np.maximum(mean_sq - media**2, 0))

        vettore_media.append(media.flatten())
        vettore_std.append(std.flatten())

    return np.concatenate(vettore_media), np.concatenate(vettore_std)

mean_channels, std_channels = [], []
for ch in canali:
    # 1. Pulizia prima
    cleaned = replace_nans_with_nearest_blockwise(globals()[ch])

    # 2. Calcolo media/std su (h,w) = (54,111) o (111,54) a seconda del tuo caso
    mean, std = media_std_mobile_immagini(cleaned, 111, 54, kernel_size=5)

    mean_channels.append(mean)
    std_channels.append(std)

    
data = {}

# Canali normali (raw) senza NaN
for i, ch in enumerate(canali):
    data[ch] = replace_nans_with_nearest_blockwise(globals()[ch].copy())

# Mascherati (terra/mare/estate/inverno) senza NaN
for i, ch in enumerate(canali):
    data[f'{ch}_mare'] = replace_nans_with_nearest_blockwise(applica_maschera(globals()[ch], Mare_maschera))
    data[f'{ch}_terra'] = replace_nans_with_nearest_blockwise(applica_maschera(globals()[ch], TERRA_maschera))
    estate = replace_nans_with_nearest_blockwise(applica_maschera(globals()[ch], STAGIONI_maschera, 0))
    inverno = replace_nans_with_nearest_blockwise(applica_maschera(globals()[ch], STAGIONI_maschera, 1))
    data[f'{ch}_estate'] = estate
    data[f'{ch}_inverno'] = inverno

# Media/std mobile già calcolate, riapplichiamo replace_nans_with_nearest per sicurezza
for i, ch in enumerate(canali):
    data[f'mean of 5x5 window CH_{i+1}'] = replace_nans_with_nearest_blockwise(mean_channels[i].copy())
for i, ch in enumerate(canali):
    data[f'std of 5x5 window CH_{i+1}'] = replace_nans_with_nearest_blockwise(std_channels[i].copy())

#for col in df_features.columns:
   #df_features[col] = replace_nans_with_nearest(df_features[col].values)
    
#df_features = df_features.fillna(0)
    
# Verifica
#print("Numero di NaN totali nel DataFrame dopo sostituzione:", df_features.isna().sum().sum())

# Creazione DataFrame e salvataggio
df_features = pd.DataFrame(data)

# === Salvataggio a blocchi ===
blocchi_size = 599_400  # numero di righe per ogni pickle
num_blocchi = len(df_features) // blocchi_size + 1

for i in range(num_blocchi):
    start = i * blocchi_size
    end = min((i + 1) * blocchi_size, len(df_features))
    df_blocco = df_features.iloc[start:end]
    nome_file = os.path.join(output_dir, f'df_features_blocco_{i+1}.pickle')
    with open(nome_file, 'wb') as f:
        pickle.dump(df_blocco, f)
    print(f"✅ Salvato blocco {i+1}/{num_blocchi}, righe {start}:{end}, file: {nome_file}")

# Salvataggio DPR e maschere separate

with open(os.path.join(output_dir, 'maschera_terra_ml.pickle'), 'wb') as f:
    pickle.dump(TERRA_maschera, f)

with open(os.path.join(output_dir, 'maschera_mare_ml.pickle'), 'wb') as f:
    pickle.dump(Mare_maschera, f)

with open(os.path.join(output_dir, 'maschera_stagioni_ml.pickle'), 'wb') as f:
    pickle.dump(STAGIONI_maschera, f)

print("✅ Tutte le feature, DPR e maschere salvate correttamente.")
```

```python
################# MAPPE ACCUMULI ORARI #####################

import pickle
import numpy as np
import matplotlib.pyplot as plt
import os
import cartopy.crs as ccrs
import cartopy.feature as cfeature
import matplotlib
import pandas as pd
from collections import defaultdict

output_file = r"D:\Enrico\y_pred.npy"

# === Percorsi ===
dir_lat = r"D:\Enrico\emiliaromagna\Dati_processati\lat.pickle"
dir_lon = r"D:\Enrico\emiliaromagna\Dati_processati\lon.pickle"
dir_time = r"D:\Enrico\emiliaromagna\Dati_processati\TIME.pickle"
output_dir = r"D:\Enrico\Mappe_SEVIRI_accumulo_orario"
os.makedirs(output_dir, exist_ok=True)

# === Carico lat, lon e time ===
with open(dir_lat, "rb") as f:
    lat_all = pickle.load(f)
with open(dir_lon, "rb") as f:
    lon_all = pickle.load(f)
with open(dir_time, "rb") as f:
    time_all = pickle.load(f)

y_pred = np.load(output_file)

# === Parametri immagine ===
img_h, img_w = 54, 111
pixels_per_image = img_h * img_w
num_images = len(y_pred) // pixels_per_image

# === Definizione intervalli e tassi rappresentativi ===
rain_intervals = [
    (0, 0.1), (0.1, 1), (1, 5), (5, 15), (15, np.inf)
]
class_to_rate = []
for low, high in rain_intervals:
    if np.isinf(high):
        val = low * 1.5
    else:
        val = (low + high) / 2.0
    class_to_rate.append(val)

def classify_precip(precip):
    """Riclassifica un campo di accumulo orario nei bin definiti"""
    classified = np.full_like(precip, fill_value=-1, dtype=int)
    for i, (low, high) in enumerate(rain_intervals):
        mask = (precip >= low) & (precip < high)
        classified[mask] = i
    return classified

# === Raggruppo per ora ===
hour_groups = defaultdict(list)
for i in range(num_images):
    t_pixel = time_all[i * pixels_per_image]
    dt = pd.to_datetime(t_pixel)
    hour_key = dt.replace(minute=0, second=0, microsecond=0)  # solo ora
    hour_groups[hour_key].append(i)

# === Colormap fissa ===
colors = ['#FFFFFF', '#ADD8E6', '#7CFC00', '#FFD700', '#8A2BE2']
cmap = matplotlib.colors.ListedColormap(colors, N=5)

print(f"✅ Gruppi orari trovati: {len(hour_groups)}\n")

# === Ciclo su ciascuna ora ===
for hour_key in sorted(hour_groups.keys()):
    img_indices = hour_groups[hour_key]
    print(f"🕒 Elaboro ora {hour_key}, immagini: {len(img_indices)}")

    # Accumulo totale (in mm) per ogni pixel
    accum_1h = np.zeros(pixels_per_image, dtype=float)

    for idx in img_indices:
        start = idx * pixels_per_image
        end = (idx+1) * pixels_per_image
        cls_map = y_pred[start:end]

        # Mappa classe -> tasso mm/h
        rates = np.array([class_to_rate[c] if c >= 0 else 0 for c in cls_map])
        # Accumulo 15'
        accum_1h += rates * 0.25  # mm

    # Riclassificazione dell'accumulo orario
    hourly_classes = classify_precip(accum_1h)

    # Reshape in 2D
    img_data = hourly_classes.reshape(img_w, img_h)
    lat_2d = lat_all[:pixels_per_image].reshape(img_h, img_w)
    lon_2d = lon_all[:pixels_per_image].reshape(img_h, img_w)

    # === Plot ===
    fig = plt.figure(figsize=(8,6))
    ax = plt.axes(projection=ccrs.PlateCarree())

    lat_min, lat_max = 43.4, 45.8
    lon_min, lon_max = 9.15, 13.2
    ax.set_extent([lon_min, lon_max, lat_min, lat_max], crs=ccrs.PlateCarree())

    im = ax.imshow(img_data.T, origin='lower', cmap=cmap,
               extent=[lon_2d.min(), lon_2d.max(), lat_2d.min(), lat_2d.max()],
               transform=ccrs.PlateCarree(), vmin=0, vmax=4)

    # Griglie
    xticks = np.linspace(lon_min, lon_max, 5)
    yticks = np.linspace(lat_min, lat_max, 5)
    ax.set_xticks(xticks)
    ax.set_yticks(yticks)

    ax.add_feature(cfeature.COASTLINE)
    ax.add_feature(cfeature.BORDERS, linestyle=':')
    ax.add_feature(cfeature.LAND, facecolor='lightgray', alpha=0.3)

    # Colorbar
    cbar = plt.colorbar(im, ax=ax, orientation='vertical', pad=0.02, shrink=0.7)
    cbar.set_label('Classe precipitazione oraria', fontsize=10)
    cbar.set_ticks([0,1,2,3,4])
    cbar.set_ticklabels(['Dry', 'Light', 'Moderate', 'Heavy', 'Intense'])
    cbar.ax.tick_params(labelsize=8)

    ax.set_xlabel('Longitudine', fontsize=12)
    ax.set_ylabel('Latitudine', fontsize=12)
    ax.set_title(f"Accumulo orario - {hour_key.strftime('%Y-%m-%d %H:00')}", fontsize=12)

    save_path = os.path.join(output_dir, f"mappa_oraria_{hour_key.strftime('%Y%m%d_%H')}.png")
    plt.tight_layout()
    plt.savefig(save_path)
    plt.close()
    print(f"✅ Mappa salvata: {save_path}\n")

print("🎉 Tutte le mappe orarie predette elaborate correttamente!")
```

```python
################# MAPPE ACCUMULI ORARI #####################

import pickle
import numpy as np
import matplotlib.pyplot as plt
import os
import cartopy.crs as ccrs
import cartopy.feature as cfeature
import matplotlib
import pandas as pd
from collections import defaultdict

output_file = r"D:\Enrico\definitivo_actual\y_pred.npy"

# === Percorsi ===
dir_lat = r"D:\Enrico\definitivo_actual\Dati_processati\lat.pickle"
dir_lon = r"D:\Enrico\definitivo_actual\Dati_processati\lon.pickle"
dir_time = r"D:\Enrico\definitivo_actual\Dati_processati\TIME.pickle"
output_dir = r"D:\Enrico\definitivo_actual\Mappe_SEVIRI_accumulo_orario_estesa"
dir_giorno = r"D:\Enrico\definitivo_actual\Dati_processati\giorno_maschera_70.pickle"

os.makedirs(output_dir, exist_ok=True)

# === Carico lat, lon e time ===
with open(dir_lat, "rb") as f:
    lat_all = pickle.load(f)
with open(dir_lon, "rb") as f:
    lon_all = pickle.load(f)
with open(dir_time, "rb") as f:
    time_all = pickle.load(f)
    
with open(dir_giorno, "rb") as f:
    GIORNO_maschera = pickle.load(f)
    
mask_indices_bal = np.where(GIORNO_maschera == 1)[0]
time_all = np.array(time_all)[mask_indices_bal]


y_pred = np.load(output_file)

# === Parametri immagine ===
img_h, img_w = 55, 112
pixels_per_image = img_h * img_w
num_images = len(y_pred) // pixels_per_image

# === Definizione intervalli e tassi rappresentativi ===
rain_intervals = [
    (0, 0.1), (0.1, 1), (1, 5), (5, 15), (15, np.inf)
]
class_to_rate = []
for low, high in rain_intervals:
    if np.isinf(high):
        val = low * 1.5
    else:
        val = (low + high) / 2.0
    class_to_rate.append(val)

def classify_precip(precip):
    """Riclassifica un campo di accumulo orario nei bin definiti"""
    classified = np.full_like(precip, fill_value=-1, dtype=int)
    for i, (low, high) in enumerate(rain_intervals):
        mask = (precip >= low) & (precip < high)
        classified[mask] = i
    return classified

# === Raggruppo per ora ===
hour_groups = defaultdict(list)
for i in range(num_images):
    t_pixel = time_all[i * pixels_per_image]
    dt = pd.to_datetime(t_pixel)
    hour_key = dt.replace(minute=0, second=0, microsecond=0)  # solo ora
    hour_groups[hour_key].append(i)

# === Colormap fissa ===
colors = [
    '#FFFFFF',  # 0 Dry - bianco
    '#A4D3EE',  # 1 Light blue - pioggia debole
    '#32CD32',  # 2 LimeGreen - pioggia moderata
    '#FFD700',  # 3 Gold - pioggia forte
    '#8A2BE2'   # 4 Violet - pioggia molto intensa
]

#colors = ['#FFFFFF', '#ADD8E6', '#7CFC00', '#FFD700', '#8A2BE2']
cmap = matplotlib.colors.ListedColormap(colors, N=5)

print(f"✅ Gruppi orari trovati: {len(hour_groups)}\n")

# === Ciclo su ciascuna ora ===
for hour_key in sorted(hour_groups.keys()):
    img_indices = hour_groups[hour_key]
    print(f"🕒 Elaboro ora {hour_key}, immagini: {len(img_indices)}")

    # Accumulo totale (in mm) per ogni pixel
    accum_1h = np.zeros(pixels_per_image, dtype=float)

    for idx in img_indices:
        start = idx * pixels_per_image
        end = (idx+1) * pixels_per_image
        cls_map = y_pred[start:end]

        # Mappa classe -> tasso mm/h
        rates = np.array([class_to_rate[c] if c >= 0 else 0 for c in cls_map])
        # Accumulo 15'
        accum_1h += rates * 0.25  # mm

    # Riclassificazione dell'accumulo orario
    hourly_classes = classify_precip(accum_1h)

    # Reshape in 2D
    img_data = hourly_classes.reshape(img_h, img_w)
    lat_2d = lat_all[:pixels_per_image].reshape(img_h, img_w)
    lon_2d = lon_all[:pixels_per_image].reshape(img_h, img_w)
    
    img_data = np.fliplr(img_data)
    #lon_2d = np.fliplr(lon_2d)

    # === Plot ===
    fig = plt.figure(figsize=(8,6))
    ax = plt.axes(projection=ccrs.PlateCarree())

    lat_min, lat_max = 43.4, 45.8
    lon_min, lon_max = 9.15, 13.2
    ax.set_extent([lon_2d.min(), lon_2d.max(), lat_2d.min(), lat_2d.max()], crs=ccrs.PlateCarree())

    im = ax.imshow(img_data, origin='lower', cmap=cmap,
               extent=[lon_2d.min(), lon_2d.max(), lat_2d.min(), lat_2d.max()],
               transform=ccrs.PlateCarree(), vmin=0, vmax=4)

    # Griglie
    xticks = np.linspace(lon_min, lon_max, 5)
    yticks = np.linspace(lat_min, lat_max, 5)
    ax.set_xticks(xticks)
    ax.set_yticks(yticks)

    ax.add_feature(cfeature.COASTLINE)
    ax.add_feature(cfeature.BORDERS, linestyle=':')
    ax.add_feature(cfeature.LAND, facecolor='lightgray', alpha=0.3)

    # Colorbar
    cbar = plt.colorbar(im, ax=ax, orientation='vertical', pad=0.04, shrink=0.5)
    cbar.set_label('Hourly precipitation classes', fontsize=14)
    cbar.set_ticks([0.5, 1.5, 2.5, 3.5, 4.5])
    cbar.set_ticklabels(['Dry', 'Light', 'Moderate', 'Heavy', 'Intense'])
    cbar.ax.tick_params(labelsize=11)

    ax.set_xlabel('Longitudine', fontsize=18)
    ax.set_ylabel('Latitudine', fontsize=18)
    #ax.set_title(f"Accumulo orario - {hour_key.strftime('%Y-%m-%d %H:00')}", fontsize=12)

    ax.tick_params(labelsize=15, direction='out', length=3, width=0.8)
    ax.xaxis.set_tick_params(pad=5)
    ax.yaxis.set_tick_params(pad=5)

    save_path = os.path.join(output_dir, f"mappa_oraria_{hour_key.strftime('%Y%m%d_%H')}.png")
    plt.tight_layout()
    plt.savefig(save_path)
    plt.close()
    print(f"✅ Mappa salvata: {save_path}\n")

print("🎉 Tutte le mappe orarie predette elaborate correttamente!")
```
