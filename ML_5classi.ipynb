```python
######### MODELLO 5 CLASSI PROVA senza maschera giorno ########

import pickle
import pandas as pd
import os
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import (
    classification_report,
    accuracy_score,
    precision_score,
    recall_score,
    f1_score,
    confusion_matrix,
    cohen_kappa_score
)

# === Definizione delle cartelle ===
directory_features = 'C:/Users/user/Documents/TESI/datigrezzi/DATI_grezzi_2020/features_5x_introno3x'
directory_features_bil5 = os.path.join(directory_features, 'bilanciamento_5classi')
output_directory_modelli_new = os.path.join(directory_features, 'output_modelli')
output_directory_new = os.path.join(directory_features, 'output')

os.makedirs(output_directory_modelli_new, exist_ok=True)
os.makedirs(output_directory_new, exist_ok=True)

# === Caricamento dei dati ===
with open(os.path.join(directory_features, 'df_features.pickle'), 'rb') as f:
    df_features = pickle.load(f)

with open(os.path.join(directory_features, 'DPR_ml.pickle'), 'rb') as f:
    DPR = pickle.load(f)

with open(os.path.join(directory_features_bil5, 'df_features.pickle'), 'rb') as f:
    df_features_balanced = pickle.load(f)

with open(os.path.join(directory_features_bil5, 'DPR_ml.pickle'), 'rb') as f:
    DPR_balanced = pickle.load(f)

with open(os.path.join(directory_features_bil5, 'flag_train.pickle'), 'rb') as f:
    flag_train = pickle.load(f)

print(f"Lunghezza df_features: {len(df_features)}")
print(f"Lunghezza DPR: {len(DPR)}")

# === Classi e soglie ===
max_value = np.max(DPR)
ordered_classes = ['Dry', 'Light', 'Moderate', 'Heavy', 'Intense']
class_ranges = [0, 0.1, 1, 5, 15, max_value]
class_to_index = {label: i for i, label in enumerate(ordered_classes)}

def map_to_class(DPR_value):
    for i in range(len(class_ranges) - 1):
        if class_ranges[i] <= DPR_value < class_ranges[i + 1]:
            return ordered_classes[i]
    return ordered_classes[-1]

# Mapping delle classi
df_features['Precipitation_Class'] = [map_to_class(val) for val in DPR]
df_features_balanced['Precipitation_Class'] = [map_to_class(val) for val in DPR_balanced]

# === Encoding manuale ===
y_train_balanced = df_features_balanced['Precipitation_Class'].map(class_to_index).values
y_test = df_features['Precipitation_Class'].map(class_to_index).values

# === Train/Test feature split ===
X_train_balanced = df_features_balanced.drop(columns=['Precipitation_Class'])
X_test = df_features.drop(columns=['Precipitation_Class'])

# === Normalizzazione ===
x_min_vals = X_train_balanced.min()
x_max_vals = X_train_balanced.max()
X_train_norm = (X_train_balanced - x_min_vals) / (x_max_vals - x_min_vals) * 2 - 1
X_test_norm = (X_test - x_min_vals) / (x_max_vals - x_min_vals) * 2 - 1

# === Modello Random Forest ===
rf_classifier = RandomForestClassifier(
    n_jobs=7,
    n_estimators=100,
    max_depth=50,
    min_samples_leaf=2,
    max_features=0.5,
    max_samples=0.5,
)
rf_classifier.fit(X_train_norm, y_train_balanced)

# === Valutazione Training ===
y_train_pred = rf_classifier.predict(X_train_norm)
print("\n=========== VALUTAZIONE SUL TRAINING SET ===========")
print(f'Accuracy (Train): {accuracy_score(y_train_balanced, y_train_pred):.4f}')
print(f'Precision (Train): {precision_score(y_train_balanced, y_train_pred, average="weighted"):.4f}')
print(f'Recall (Train): {recall_score(y_train_balanced, y_train_pred, average="weighted"):.4f}')
print(f'F1-Score (Train): {f1_score(y_train_balanced, y_train_pred, average="weighted"):.4f}')

# === Valutazione Test ===
y_pred = rf_classifier.predict(X_test_norm)
y_test_masked = np.array(y_test)[~flag_train]
y_pred_masked = np.array(y_pred)[~flag_train]

accuracy = accuracy_score(y_test_masked, y_pred_masked)
precision = precision_score(y_test_masked, y_pred_masked, average='weighted')
recall = recall_score(y_test_masked, y_pred_masked, average='weighted')
f1 = f1_score(y_test_masked, y_pred_masked, average='weighted')

conf_matrix = confusion_matrix(y_test_masked, y_pred_masked, labels=[0, 1, 2, 3, 4])

# === Metriche per classe ===
POD_list = []
FAR_list = []
BIAS_list = []
CSI_list = []
ETS_list = []

for i, class_name in enumerate(ordered_classes):
    TP = conf_matrix[i, i]
    FP = conf_matrix[:, i].sum() - TP
    FN = conf_matrix[i, :].sum() - TP

    POD = TP / (TP + FN) if (TP + FN) > 0 else 0
    FAR = FP / (TP + FP) if (TP + FP) > 0 else 0
    BIAS = (TP + FP) / (TP + FN) if (TP + FN) > 0 else 0
    CSI = TP / (TP + FP + FN) if (TP + FP + FN) > 0 else 0
    ETS = (TP - ((TP + FP) * (TP + FN) / conf_matrix.sum())) / \
          (TP + FP + FN - ((TP + FP) * (TP + FN) / conf_matrix.sum())) if (TP + FP + FN) > 0 else 0

    POD_list.append(POD)
    FAR_list.append(FAR)
    BIAS_list.append(BIAS)
    CSI_list.append(CSI)
    ETS_list.append(ETS)

    print(f"\nClass: {class_name}")
    print(f"POD: {POD:.3f}, FAR: {FAR:.3f}, BIAS: {BIAS:.3f}, CSI: {CSI:.3f}, ETS: {ETS:.3f}")

# === Medie delle metriche ===
print("\n=== VALORI MEDI DELLE METRICHE ===")
print(f"Mean POD:  {np.mean(POD_list):.3f}")
print(f"Mean FAR:  {np.mean(FAR_list):.3f}")
print(f"Mean BIAS: {np.mean(BIAS_list):.3f}")
print(f"Mean CSI:  {np.mean(CSI_list):.3f}")
print(f"Mean ETS:  {np.mean(ETS_list):.3f}")

# === METRICHE BINARIE: PIOGGIA (1–4) VS NON PIOGGIA (0) ===
# Riclassificazione binaria
y_test_bin = np.where(y_test_masked == 0, 0, 1)
y_pred_bin = np.where(y_pred_masked == 0, 0, 1)

# Confusione binaria: [[TN, FP], [FN, TP]]
cm_bin = confusion_matrix(y_test_bin, y_pred_bin, labels=[0, 1])
TN, FP, FN, TP = cm_bin[0,0], cm_bin[0,1], cm_bin[1,0], cm_bin[1,1]

# Calcolo metriche
POD  = TP / (TP + FN) if (TP + FN) > 0 else 0
FAR  = FP / (TP + FP) if (TP + FP) > 0 else 0
BIAS = (TP + FP) / (TP + FN) if (TP + FN) > 0 else 0
CSI  = TP / (TP + FP + FN) if (TP + FP + FN) > 0 else 0

ETS_denom = TP + FP + FN - ((TP + FP) * (TP + FN) / (TP + FP + FN + TN))
ETS  = (TP - ((TP + FP) * (TP + FN) / (TP + FP + FN + TN))) / ETS_denom if ETS_denom > 0 else 0

# Stampa risultati binari
print("\n=========== METRICHE BINARIE: PIOGGIA VS NON PIOGGIA ===========")
print(f"POD:  {POD:.3f}")
print(f"FAR:  {FAR:.3f}")
print(f"BIAS: {BIAS:.3f}")
print(f"CSI:  {CSI:.3f}")
print(f"ETS:  {ETS:.3f}")


# === HSS e report ===
def hss_empirical(y_true, y_pred):
    cm = confusion_matrix(y_true, y_pred)
    total = cm.sum()
    correct = np.trace(cm)
    expected = np.sum((cm.sum(axis=1) * cm.sum(axis=0)) / total)
    return (correct - expected) / (total - expected) if (total - expected) != 0 else 0

hss_value = hss_empirical(y_test_masked, y_pred_masked)
hss_1 = cohen_kappa_score(y_test_masked, y_pred_masked)

print("\n=========== VALUTAZIONE SUL TEST SET ===========")
print(f"HSS empirica: {hss_value:.3f}")
print(f"Heidke Skill Score (CK): {hss_1:.3f}")
print(f'Accuracy (Test): {accuracy:.4f}')
print(f'Precision (Test): {precision:.4f}')
print(f'Recall (Test): {recall:.4f}')
print(f'F1-Score (Test): {f1:.4f}')
print('\nConfusion Matrix:')
print(conf_matrix)

print('\nClassification Report:')
print(classification_report(y_test_masked, y_pred_masked, target_names=ordered_classes))
```

```python
######### MODELLO 5 CLASSI PROVA senza maschera giorno + selezione feature ########

import pickle
import pandas as pd
import os
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import (
    classification_report,
    accuracy_score,
    precision_score,
    recall_score,
    f1_score,
    confusion_matrix,
    cohen_kappa_score
)

# === Definizione delle cartelle ===
directory_features = 'C:/Users/user/Documents/TESI/datigrezzi/DATI_grezzi_2020/Dati_ML_plus_5classi'
directory_features_bil5 = os.path.join(directory_features, 'bilanciamento_5classi')
output_directory_modelli_new = os.path.join(directory_features, 'output_modelli')
output_directory_new = os.path.join(directory_features, 'output')

os.makedirs(output_directory_modelli_new, exist_ok=True)
os.makedirs(output_directory_new, exist_ok=True)

# === Caricamento dei dati ===
with open(os.path.join(directory_features, 'df_features.pickle'), 'rb') as f:
    df_features = pickle.load(f)

with open(os.path.join(directory_features, 'DPR_ml.pickle'), 'rb') as f:
    DPR = pickle.load(f)

with open(os.path.join(directory_features_bil5, 'df_features.pickle'), 'rb') as f:
    df_features_balanced = pickle.load(f)

with open(os.path.join(directory_features_bil5, 'DPR_ml.pickle'), 'rb') as f:
    DPR_balanced = pickle.load(f)

with open(os.path.join(directory_features_bil5, 'flag_train.pickle'), 'rb') as f:
    flag_train = pickle.load(f)

print(f"Lunghezza df_features: {len(df_features)}")
print(f"Lunghezza DPR: {len(DPR)}")

# === Classi e soglie ===
max_value = np.max(DPR)
ordered_classes = ['Dry', 'Light', 'Moderate', 'Heavy', 'Intense']
class_ranges = [0, 0.1, 1, 5, 15, max_value]
class_to_index = {label: i for i, label in enumerate(ordered_classes)}

def map_to_class(DPR_value):
    for i in range(len(class_ranges) - 1):
        if class_ranges[i] <= DPR_value < class_ranges[i + 1]:
            return ordered_classes[i]
    return ordered_classes[-1]

# Mapping delle classi
df_features['Precipitation_Class'] = [map_to_class(val) for val in DPR]
df_features_balanced['Precipitation_Class'] = [map_to_class(val) for val in DPR_balanced]

# === Encoding manuale ===
y_train_balanced = df_features_balanced['Precipitation_Class'].map(class_to_index).values
y_test = df_features['Precipitation_Class'].map(class_to_index).values

# === Train/Test feature split ===
X_train_balanced = df_features_balanced.drop(columns=['Precipitation_Class'])
X_test = df_features.drop(columns=['Precipitation_Class'])

# === Normalizzazione ===
x_min_vals = X_train_balanced.min()
x_max_vals = X_train_balanced.max()
X_train_norm = (X_train_balanced - x_min_vals) / (x_max_vals - x_min_vals) * 2 - 1
X_test_norm = (X_test - x_min_vals) / (x_max_vals - x_min_vals) * 2 - 1

# === Random Forest iniziale per calcolare importanza feature ===
rf_classifier = RandomForestClassifier(
    random_state=42,
    n_jobs=7,
    n_estimators=100,
    max_depth=50,
    min_samples_leaf=2,
    max_features=0.5,
    max_samples=0.5,
)
rf_classifier.fit(X_train_norm, y_train_balanced)

# === Importanza delle feature ===
importances = rf_classifier.feature_importances_
feature_names = X_train_balanced.columns
indices = np.argsort(importances)[::-1]

# Plot delle top 30 feature
top_n = 5
plt.figure(figsize=(12, 6))
plt.title("Importanza delle feature (Top 30)")
plt.bar(range(top_n), importances[indices[:top_n]], align="center")
plt.xticks(range(top_n), [feature_names[i] for i in indices[:top_n]], rotation=90)
plt.tight_layout()
plt.show()

# Classifica completa
feature_ranking = pd.DataFrame({
    'Feature': feature_names[indices],
    'Importance': importances[indices]
})
print("\nClassifica delle feature per importanza:")
print(feature_ranking)

# === Selezione feature più importanti ===
selected_features = feature_ranking[feature_ranking['Importance'] > 0.025]['Feature'].tolist()
print(f"\nNumero di feature selezionate: {len(selected_features)}")
print("Feature selezionate:")
print(selected_features)

# Salva le feature selezionate
with open(os.path.join(output_directory_modelli_new, 'selected_features.pickle'), 'wb') as f:
    pickle.dump(selected_features, f)

# === Riduzione dei dataset ===
X_train_selected = X_train_norm[selected_features]
X_test_selected = X_test_norm[selected_features]

# === Nuovo training solo con feature selezionate ===
rf_classifier.fit(X_train_selected, y_train_balanced)

# === Valutazione Training ===
y_train_pred = rf_classifier.predict(X_train_selected)
print("\n=========== VALUTAZIONE SUL TRAINING SET ===========")
print(f'Accuracy (Train): {accuracy_score(y_train_balanced, y_train_pred):.4f}')
print(f'Precision (Train): {precision_score(y_train_balanced, y_train_pred, average="weighted"):.4f}')
print(f'Recall (Train): {recall_score(y_train_balanced, y_train_pred, average="weighted"):.4f}')
print(f'F1-Score (Train): {f1_score(y_train_balanced, y_train_pred, average="weighted"):.4f}')

# === Valutazione Test ===
y_pred = rf_classifier.predict(X_test_selected)
y_test_masked = np.array(y_test)[~flag_train]
y_pred_masked = np.array(y_pred)[~flag_train]

accuracy = accuracy_score(y_test_masked, y_pred_masked)
precision = precision_score(y_test_masked, y_pred_masked, average='weighted')
recall = recall_score(y_test_masked, y_pred_masked, average='weighted')
f1 = f1_score(y_test_masked, y_pred_masked, average='weighted')

conf_matrix = confusion_matrix(y_test_masked, y_pred_masked, labels=[0, 1, 2, 3, 4])

# === Metriche per classe ===
POD_list = []
FAR_list = []
BIAS_list = []
CSI_list = []
ETS_list = []

for i, class_name in enumerate(ordered_classes):
    TP = conf_matrix[i, i]
    FP = conf_matrix[:, i].sum() - TP
    FN = conf_matrix[i, :].sum() - TP

    POD = TP / (TP + FN) if (TP + FN) > 0 else 0
    FAR = FP / (TP + FP) if (TP + FP) > 0 else 0
    BIAS = (TP + FP) / (TP + FN) if (TP + FN) > 0 else 0
    CSI = TP / (TP + FP + FN) if (TP + FP + FN) > 0 else 0
    ETS = (TP - ((TP + FP) * (TP + FN) / conf_matrix.sum())) / \
          (TP + FP + FN - ((TP + FP) * (TP + FN) / conf_matrix.sum())) if (TP + FP + FN) > 0 else 0

    POD_list.append(POD)
    FAR_list.append(FAR)
    BIAS_list.append(BIAS)
    CSI_list.append(CSI)
    ETS_list.append(ETS)

    print(f"\nClass: {class_name}")
    print(f"POD: {POD:.3f}, FAR: {FAR:.3f}, BIAS: {BIAS:.3f}, CSI: {CSI:.3f}, ETS: {ETS:.3f}")

# === Medie delle metriche ===
print("\n=== VALORI MEDI DELLE METRICHE ===")
print(f"Mean POD:  {np.mean(POD_list):.3f}")
print(f"Mean FAR:  {np.mean(FAR_list):.3f}")
print(f"Mean BIAS: {np.mean(BIAS_list):.3f}")
print(f"Mean CSI:  {np.mean(CSI_list):.3f}")
print(f"Mean ETS:  {np.mean(ETS_list):.3f}")

# === METRICHE BINARIE: PIOGGIA (1–4) VS NON PIOGGIA (0) ===
# Riclassificazione binaria
y_test_bin = np.where(y_test_masked == 0, 0, 1)
y_pred_bin = np.where(y_pred_masked == 0, 0, 1)

# Confusione binaria: [[TN, FP], [FN, TP]]
cm_bin = confusion_matrix(y_test_bin, y_pred_bin, labels=[0, 1])
TN, FP, FN, TP = cm_bin[0,0], cm_bin[0,1], cm_bin[1,0], cm_bin[1,1]

# Calcolo metriche
POD  = TP / (TP + FN) if (TP + FN) > 0 else 0
FAR  = FP / (TP + FP) if (TP + FP) > 0 else 0
BIAS = (TP + FP) / (TP + FN) if (TP + FN) > 0 else 0
CSI  = TP / (TP + FP + FN) if (TP + FP + FN) > 0 else 0

ETS_denom = TP + FP + FN - ((TP + FP) * (TP + FN) / (TP + FP + FN + TN))
ETS  = (TP - ((TP + FP) * (TP + FN) / (TP + FP + FN + TN))) / ETS_denom if ETS_denom > 0 else 0

# Stampa risultati binari
print("\n=========== METRICHE BINARIE: PIOGGIA VS NON PIOGGIA ===========")
print(f"POD:  {POD:.3f}")
print(f"FAR:  {FAR:.3f}")
print(f"BIAS: {BIAS:.3f}")
print(f"CSI:  {CSI:.3f}")
print(f"ETS:  {ETS:.3f}")
# === HSS e report ===
def hss_empirical(y_true, y_pred):
    cm = confusion_matrix(y_true, y_pred)
    total = cm.sum()
    correct = np.trace(cm)
    expected = np.sum((cm.sum(axis=1) * cm.sum(axis=0)) / total)
    return (correct - expected) / (total - expected) if (total - expected) != 0 else 0

hss_value = hss_empirical(y_test_masked, y_pred_masked)
hss_1 = cohen_kappa_score(y_test_masked, y_pred_masked)

print("\n=========== VALUTAZIONE SUL TEST SET ===========")
print(f"HSS empirica: {hss_value:.3f}")
print(f"Heidke Skill Score (CK): {hss_1:.3f}")
print(f'Accuracy (Test): {accuracy:.4f}')
print(f'Precision (Test): {precision:.4f}')
print(f'Recall (Test): {recall:.4f}')
print(f'F1-Score (Test): {f1:.4f}')
print('\nConfusion Matrix:')
print(conf_matrix)

print('\nClassification Report:')
print(classification_report(y_test_masked, y_pred_masked, target_names=ordered_classes))
```
